```python
columns_to_keep =['job_title',
'salary_usd',
'experience_level',
'employment_type',
'company_location',
'remote_ratio',
'required_skills',
'education_required',
'years_experience',
'industry',
'company_size',
'posting_date'
]
df=df[columns_to_keep]
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for i in df:
  if (df[i].dtype=='object'):
    df[i]=le.fit_transform(df[i])
```

```python
df.drop(columns=['posting_date'], inplace=True)
```

```python
df.head()
```

```python
Target= 'salary_usd'
```

```python
num_coln = [
    'years_experience',
    'remote_ratio'
]
```

```python
X = df.drop('salary_usd', axis=1)
y = df['salary_usd']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2 , random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train[num_coln] = scaler.fit_transform(X_train[num_coln])
X_test[num_coln] = scaler.transform(X_test[num_coln])
```

# Model Selection


## Linear Regression

```python
from sklearn.linear_model import LinearRegression

li = LinearRegression()
```

```python
li.fit(X_train,y_train)
```

```python
li.intercept_
```

The value of c is 116523 --->(y= mX+c)

```python
li.coef_
```

There are 10 multiple m values

```python
y_pred = li.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
print("MAE:", mean_absolute_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R2:", r2_score(y_test, y_pred))
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import classification_report,accuracy_score

dt = DecisionTreeRegressor()
```

```python
dt.fit(X_train,y_train)
```

```python
y_pred = dt.predict(X_test)
```

```python
print("MAE:", mean_absolute_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R2:", r2_score(y_test, y_pred))
```

## RandomForest


```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
print("MAE:", mean_absolute_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R2:", r2_score(y_test, y_pred))
```

```python
df['experience_level'] = df['experience_level'].replace({
    'EN' :'Entry',
    'MI' :'Mid',
    'SE' :'Senior',
    'EX' :'Executive'
})
```

```python
df['employment_type'] = df['employment_type'].replace({
    'FT': 'Full-time',
    'PT': 'Part-time',
    'CT': 'Contract',
    'FL': 'Freelance'
})
```

```python
df['remote_ratio']=df['remote_ratio'].replace({
    0 :'No remote',
    50: 'Hybrid',
    100 :'Fully remote'
})

df['company_size'] = df['company_size'].replace({
    'M' : 'Mid(50-250)',
    'S' : 'Small(<50)',
    'L' : 'Large(>250)'
})
```
